```python
import sys
import pygame
from pygame.locals import *

#Racing Wheel  .
class Racing_Wheel:
    '''
      (pygame)       .
      .
     ,    .
    
    Init_Racing_Wheel() ->  ,   , clock, joystick 
    '''

    def __init__(self, clock = 0, joysticks = 0):   
        self.clock, self.joysticks = self.Init_Racing_Wheel()
    
    def test():
        print("Nothing Happen")
    
    #Racing Wheel    , clock joystick 
    def Init_Racing_Wheel(self):
        pygame.init()
        pygame.joystick.init()
        self.clock = pygame.time.Clock()
        self.joysticks = [pygame.joystick.Joystick(i) for i in range(pygame.joystick.get_count())]
        for joystick in self.joysticks:
            print(joystick.get_name())
        return self.clock, self.joysticks
    
    #Wheel Input  ,     
    def Input_Wheel(self):
        temp_square_color = 0
        temp_motion = [0,0]
        temp_colors = [(255, 0, 0), (0, 255, 0), (0, 0, 255)]
        for event in pygame.event.get():
            if event.type == pygame.JOYBUTTONDOWN:
                print(event)
                if event.button == 0:
                    temp_square_color = (temp_square_color + 1) % len(temp_colors)
            if event.type == pygame.JOYBUTTONUP:
                print(event)
            if event.type == pygame.JOYAXISMOTION:
                print(event)
                if event.axis < 2:
                    temp_motion[event.axis] = event.value
            if event.type == pygame.JOYHATMOTION:
                print(event)
            if event.type == pygame.JOYDEVICEADDED:
                joysticks = [pygame.joystick.Joystick(i) for i in range(pygame.joystick.get_count())]
                for joystick in joysticks:
                    print(joystick.get_name())
            if event.type == pygame.JOYDEVICEREMOVED:
                joysticks = [pygame.joystick.Joystick(i) for i in range(pygame.joystick.get_count())]
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_ESCAPE:
                    pygame.quit()
                    sys.exit()
        return temp_square_color, temp_motion, temp_colors

#Pygame Test .
class Test_Pygame:
    '''
       .
       ,       .
    
    Test_Pygame_Monitor_Set() ->     ,      
    ->       
    '''
    
    def __init__(self, screen = 0, my_square = 0, my_square_color = 0, colors = 0, motion = 0):
        self.screen, self.my_square, self.my_square_color, self.colors, self.motion = self.Test_Init_Pygame_Monitor_Set()
    
    #Test   
    def Test_Init_Pygame_Monitor_Set(self):
        pygame.display.set_caption('game base')
        self.screen = pygame.display.set_mode((500, 500), 0, 32)
        self.my_square = pygame.Rect(50, 50, 50, 50)
        self.my_square_color = 0
        self.colors = [(255, 0, 0), (0, 255, 0), (0, 0, 255)]
        self.motion = [0, 0]
        return self.screen, self.my_square, self.my_square_color, self.colors, self.motion
    
    #Test   
    def Test_Pygame_Monitor_Draw(self):
        self.screen.fill((0, 0, 0))
        pygame.draw.rect(self.screen, self.colors[self.my_square_color], self.my_square)
        if abs(self.motion[0]) < 0.1:
            self.motion[0] = 0
        if abs(self.motion[1]) < 0.1:
            self.motion[1] = 0
        self.my_square.x += self.motion[0] * 10
        self.my_square.y += self.motion[1] * 10
        
    def Test_Get_Datas(self, Raching_Wheel_Datas):
        self.my_square_color, self.motion, self.colors = Raching_Wheel_Datas
        

        
'''
MAIN
'''

#  , while  
#try:
#Init
Racing_Wheel_Test = Racing_Wheel()
Test_Monitor_N_Btn = Test_Pygame()

#Loop
while True:
    Test_Monitor_N_Btn.Test_Pygame_Monitor_Draw()
    monitor_datas = Racing_Wheel_Test.Input_Wheel()
    
    Test_Monitor_N_Btn.Test_Get_Datas(monitor_datas)
    
    pygame.display.update()
    Racing_Wheel_Test.clock.tick(60)
        
        
# except KeyboardInterrupt:
    
```
